```python
import newspaper
from newspaper import Article
import spacy
import re
from spacy import displacy
import spacy_transformers
```

```python
# Download large pipeline for Urkaine language
!python -m spacy download uk_core_news_trf
```

```python
# Link of Dnipro river from the Wikipedia
url = 'https://uk.wikipedia.org/wiki/%D0%94%D0%BD%D1%96%D0%BF%D1%80%D0%BE'

# Use Article to work with the url
article = Article(url)
article.download()
article.parse()

print("Text of article: \n", article.text)

# Load pipeline
nlp = spacy.load("uk_core_news_trf")
my_text_pp = article.text

# Parse text
my_text_pp = my_text_pp.lower()
my_text_pp = "".join([char if char.isalpha() or char.isspace() else " " for char in my_text_pp])
my_text_pp = re.sub(' +', ' ', my_text_pp)
doc = nlp(my_text_pp)
my_text_pp = [token.lemma_ for token in doc]
my_text_pp = " ".join(my_text_pp)

# Getting unique words
unique_words_pp = set(my_text_pp.split())
print(f"The text has {len(unique_words_pp)} unique words after preprocessing.")
print(unique_words_pp)

doc_trf = nlp(my_text_pp)

# Helper func to calculate NER
def count_ner_types(doc):
    ner_counts = {}
    for ent in doc.ents:
        ner_type = ent.label_
        if ner_type in ner_counts:
            ner_counts[ner_type] += 1
        else:
            ner_counts[ner_type] = 1
    return ner_counts

# Calculate NER
ner_counts_trf = count_ner_types(doc_trf)

print("\n" + "NER Counts on a preprocessed text:")
for ner_type, count in ner_counts_trf.items():
    print(f"{ner_type}: {count}")

# Selected NER
displacy.render(doc_trf, style="ent", jupyter=True)
```

## Conclusion:


I conducted a test of SpaCy’s Named Entity Recognition (NER) capabilities using a Wikipedia article about Dnipro River. For this experiment, I utilized the transformer-based model uk_core_news_trf. After preprocessing the text, the model produced impressive results, accurately identifying and marking all entities. This performance highlights the strength of transformer models in handling Ukrainian text effectively.
